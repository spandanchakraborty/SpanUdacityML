```python
import math as m
import numpy as np
from matplotlib import pyplot as plt
#import pandas as df
%matplotlib inline
plt.figure(figsize = (20,10))

a=np.arange(0.,1,.1)
print(a)
def plotexp(a=0):
    t={}
    t1=50
    while t1:
        t1=t1-1
        t[t1]=m.exp(a*t1*(-1))
    
    t4=[t[x] for x in t]
    plt.plot(t4, label=a)
    plt.legend(loc='upper right')

for x in a:
    plotexp(x)
```
