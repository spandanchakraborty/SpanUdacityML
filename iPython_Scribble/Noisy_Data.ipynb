```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
size=10
a=np.random.normal(scale=5,size=2*size)
#print(a)
b=np.round(np.reshape(a,(-1,2)),2)  # first Arg for reshape uses the no# of
#blocks, 2nd arg is no# of rows, 3rd Arg is the no# of columns
c=np.array([[np.sin(x[0])+np.sin(x[1])] for x in b])
plt.plot(c)
```

```python
def performance_metric(y_true, y_predict):
    from sklearn.metrics import r2_score
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    
    # TODO: Calculate the performance score between 'y_true' and 'y_predict'
    score = r2_score(y_true, y_predict)
    
    # Return the score
    return score
```

```python
score = performance_metric([3, -0.5, 2, 7, 4.2], [2.5, 0.0, 2.1, 7.8, 5.3])
print "Model has a coefficient of determination, R^2, of {:.3f}.".format(score)
```
